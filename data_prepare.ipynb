```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Global_Cybersecurity_Threats_2015-2024.csv")

# Data Understanding - Variable & Data Type
print(df.dtypes)
print("\n")
print(f"{len(df.columns)} Column")
 
# Data Preparation - Handle Missing Value
df.isnull().sum()

# Data Preparation - Invalid Numeric Column
ratio_column = ['Year','Financial Loss (in Million $)','Number of Affected Users','Incident Resolution Time (in Hours)']
for dt in ratio_column:
    plt.boxplot(df[dt])
    plt.title(f'Boxplot of {dt}')
    plt.show()
    
    invalid = df[df[dt] < 0]
    print(invalid.index)
```
