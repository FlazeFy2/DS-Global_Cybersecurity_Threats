```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Exploratory Data Analysis (EDA) - Tree Map Total Affected Users per Country
# Dataset
data = pd.read_csv("Global_Cybersecurity_Threats_2015-2024.csv")

country_affected_users = data.groupby('Country')['Number of Affected Users'].sum().reset_index()

# Sort 
country_affected_users = country_affected_users.sort_values(by='Number of Affected Users', ascending=False)

# Extract
sizes = country_affected_users['Number of Affected Users']
labels = country_affected_users.apply(lambda x: f"{x['Country']}\n({x['Number of Affected Users']})", axis=1)

# Plot Tree Map
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.title("Total Affected Users per Country", fontsize=14)
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Exploratory Data Analysis (EDA) - Tree Map Total Cyber Attack per Country
# Dataset
data = pd.read_csv("Global_Cybersecurity_Threats_2015-2024.csv")

total_attack = data.groupby('Country')['Number of Affected Users'].count().reset_index()

# Sort 
total_attack = total_attack.sort_values(by='Number of Affected Users', ascending=False)

# Extract
sizes = total_attack['Number of Affected Users']
labels = total_attack.apply(lambda x: f"{x['Country']}\n({x['Number of Affected Users']})", axis=1)

# Plot Tree Map
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.title("Total Cyber Attack per Country", fontsize=14)
plt.axis('off')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Exploratory Data Analysis (EDA) - Tree Map Total Average Affected User per Cyber Attack by Country
# Dataset
data = pd.read_csv("Global_Cybersecurity_Threats_2015-2024.csv")

# Calculate total attacks per country
total_attack = data.groupby('Country')['Number of Affected Users'].count().reset_index()
total_attack.rename(columns={'Number of Affected Users': 'Total Attacks'}, inplace=True)

# Calculate total affected users per country
country_affected_users = data.groupby('Country')['Number of Affected Users'].sum().reset_index()

# Merge both datasets
merged_data = total_attack.merge(country_affected_users, on='Country')

# Calculate average 
merged_data['Avg Affected Users'] = (merged_data['Number of Affected Users'] / merged_data['Total Attacks']).round(2)
# Sort 
merged_data = merged_data.sort_values(by='Total Attacks', ascending=False)

# Extract values
sizes = merged_data['Total Attacks']
labels = merged_data.apply(lambda x: f"{x['Country']}\nAttacks: {x['Total Attacks']}\nAvg Affected: {x['Avg Affected Users']}", axis=1)

# Plot Tree Map
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.title("Total Average Affected User per Cyber Attack by Country", fontsize=14)
plt.axis('off')  
plt.show()
```
