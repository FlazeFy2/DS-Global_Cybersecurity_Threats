```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Exploratory Data Analysis (EDA) - Tree Map Total Affected Users per Country
# Dataset
data = pd.read_csv("Global_Cybersecurity_Threats_2015-2024.csv")

country_affected_users = data.groupby('Country')['Number of Affected Users'].sum().reset_index()

# Sort 
country_affected_users = country_affected_users.sort_values(by='Number of Affected Users', ascending=False)

# Extract
sizes = country_affected_users['Number of Affected Users']
labels = country_affected_users.apply(lambda x: f"{x['Country']}\n({x['Number of Affected Users']})", axis=1)

# Plot Tree Map
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=0.8)
plt.title("Total Affected Users per Country", fontsize=14)
plt.axis('off')
plt.show()
```
